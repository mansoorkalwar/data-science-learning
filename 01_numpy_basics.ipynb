>  ### Creating Arrays

```python
import numpy as np

array = np.array([23,34,45,56,65])
print(array)
```


> ### Array Attributes

```python
arr = np.array(
                [[2,1,3,2],
                [3,4,5,6]]
            )

print(arr)
print(f"Array shape: {arr.shape}")
print(f"Array size: {arr.size}")
print(f"Array diamensions: {arr.ndim}")
print(f"Array elements type: {arr.dtype}")
```
